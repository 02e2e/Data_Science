```python
import os
os.listdir(".")
```

```python
targets=[]
names = []
for root, dirs, files in os.walk("."):
    for file in files:
        if "ipynb" in file:
            pass
        else:
            names.append(os.path.join(root,file))
            if "spam" in root:
                targets.append(1)
            else:
                targets.append(0)

```

```python
import pandas as pd
data = pd.DataFrame({"File":names,"target":targets})
```

```python
data
```

```python
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
X=data['data']
```

```python
X
```

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=10)
cluster.fit(X)
```

```python
cluster.predict(X)
```

```python
data['target']
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(data['target'],cluster.predict(X))
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2)
model.fit(X)
```
